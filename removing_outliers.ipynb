# Data Visualization

1. Count Plot & Box Plot

```python
# outcome count plot
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_all_columns(df):
    num_cols = df.shape[1]-1
    fig, axes = plt.subplots(num_cols, 2, figsize=(12, 5*num_cols))
    
    for i, col_name in enumerate(df.columns[:-1]):
        sns.histplot(data=df, x=col_name, ax=axes[i, 0])
        sns.boxplot(data=df, y=col_name, ax=axes[i, 1])
        axes[i, 0].set_title(f'{col_name} - Histogram')
        axes[i, 1].set_title(f'{col_name} - Boxplot')
        axes[i, 0].set_ylabel('')
        axes[i, 1].set_ylabel('')
    
    plt.tight_layout()

# Example usage
visualize_all_columns(df)
plt.show()
```

```python
try:
    #for figure having columns
    fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20,10))

    # Flatten the axes array to easily iterate over it
    axes = axes.flatten()

    # Loop through each column and plot a boxplot on each subplot
    for i, column in enumerate(df.columns):
        df[column].plot(kind='box', ax=axes[i])
        axes[i].set_title(column)

    
    plt.tight_layout()

    
    plt.show()

except IndexError:
    print("Number of columns in the DataFrame doesn't match the number of subplots requested.")
```

# IQR & Capping - to remove outliers

```python
def visualize_with_capping(df):
    capped_df = df.copy()
    for col in capped_df.columns[:-1]:
        q1 = capped_df[col].quantile(0.25)
        q3 = capped_df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        capped_df[col] = capped_df[col].clip(lower=lower_bound, upper=upper_bound)
    

    num_cols = capped_df.shape[1] - 1  
    fig, axes = plt.subplots(num_cols, 2, figsize=(12, 5*num_cols))
    
    for i, col_name in enumerate(capped_df.columns[:-1]):  
        sns.histplot(data=capped_df, x=col_name, ax=axes[i, 0])
        sns.boxplot(data=capped_df, y=col_name, ax=axes[i, 1])
        axes[i, 0].set_title(f'{col_name} - Histogram')
        axes[i, 1].set_title(f'{col_name} - Boxplot')
        axes[i, 0].set_ylabel('')
        axes[i, 1].set_ylabel('')
    
    plt.tight_layout()

visualize_with_capping(df)
plt.show()
```

# Split the data into X & Y

```python
target = 'Outcome'

y = df[target]

x = df.drop(target , axis=1)
```

```python
x.head()
```

```python
y.head()
```

# Apply Feature Scaling

```python
# Apply standar scaler
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
scaler.fit(x)
ssx = scaler.transform(x)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(ssx, y, test_size=0.2, random_state=7)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

# Build the Classification Algorithm

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver = 'liblinear',multi_class= 'ovr')
lr.fit(x_train,y_train)
```

```python
# K-Nearest neighbors classifier KNN

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
# Naive-Bayes Classifier

from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
```

```python
# Support Vector Machine SVM

from sklearn.svm import SVC
sv=SVC()
sv.fit(x_train,y_train)
```

```python
# Decission Tress Classifier

from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
# Random Forest 

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(criterion = 'entropy')
rf.fit(x_train,y_train)
```

# Making Predictions

```python
# making prediction on test data using logistic regression
lr_pred = lr.predict(x_test)

# making prediction on test data using KNN
knn_pred = knn.predict(x_test)

# making prediction on test data using Naive Bayes
nb_pred = nb.predict(x_test)

# making prediction on test data using SVM
sv_pred =sv.predict(x_test)

# making prediction on test data using Decision Tree
dt_pred = dt.predict(x_test)

# making prediction on test data using Random forest
rf_pred = rf.predict(x_test)
```

# Model Evalution

Train Score & Test Score

# Confusion Matrix

Confusion matrix is a table which is used to describe the performance of a classification problem.

It visualize the accuracy of a classifier by comparing predicted values with actual values.

the terms in cm:

__True Postive__: The predicted result is +ve, while it is labelled as +ve

__False Positive__:The predicted result is +ve, while it is labelled as -ve

__False Negative__:The predicted result is -ve, while it is labelled as +ve

__True Negative__:The predicted result is -ve, while it is labelled as -ve

```python
from sklearn.metrics import classification_report , confusion_matrix
cm = confusion_matrix(y_test,lr_pred)
cm
```

```python
sns.heatmap(confusion_matrix(y_test,lr_pred),annot=True,fmt="d")
```
